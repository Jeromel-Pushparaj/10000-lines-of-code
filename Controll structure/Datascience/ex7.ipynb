```python
# prompt: Reading data from text files, Excel and the web and exploring various commands for doing  descriptive analytics on the Iris data set

import pandas as pd
import matplotlib.pyplot as plt


# Load the Iris dataset from a URL
iris_data = pd.read_csv('/content/iris.csv')

# Explore the Iris dataset
print(iris_data.head())
print(iris_data.describe())
print(iris_data.info())

# Calculate the mean, median, and standard deviation of each column
print(iris_data.mean())
print(iris_data.median())
print(iris_data.std())

# Calculate the correlation matrix
print(iris_data.corr())

# Create a scatter plot of sepal length vs. sepal width
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# prompt: Use the diabetes data set from Pima Indians Diabetes data set for performing the following: Apply Univariate analysis: • Frequency • Mean, • Median, • Mode, • Variance • Standard Deviation • Skewness and Kurtosis

import pandas as pd

# Load the diabetes dataset
diabetes_data = pd.read_csv("/content/drive/MyDrive/diabetes.csv")

# Frequency
print(diabetes_data["Outcome"].value_counts())

# Mean
print(diabetes_data["Outcome"].mean())

# Median
print(diabetes_data["Outcome"].median())

# Mode
print(diabetes_data["Outcome"].mode())

# Variance
print(diabetes_data["Outcome"].var())

# Standard deviation
print(diabetes_data["Outcome"].std())

# Skewness
print(diabetes_data["Outcome"].skew())

# Kurtosis
print(diabetes_data["Outcome"].kurtosis())
```

```python
# prompt: Use the diabetes data set from Pima Indians Diabetes data set for performing the following: Apply Bivariate analysis: • Linear and logistic regression modeling

# Import necessary libraries
import statsmodels.api as sm

# Define the independent and dependent variables
X = diabetes_data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = diabetes_data['Outcome']

# Fit a linear regression model
linear_model = sm.OLS(y, sm.add_constant(X)).fit()

# Print the summary of the linear regression model
print(linear_model.summary())

# Fit a logistic regression model
logistic_model = sm.Logit(y, sm.add_constant(X)).fit()

# Print the summary of the logistic regression model
print(logistic_model.summary())
```

```python
# prompt: Use the diabetes data set from Pima Indians Diabetes data set for performing the following: Apply Bivariate analysis: • Multiple Regression analysis

# Import necessary libraries
import pandas as pd
import statsmodels.api as sm

# Load the diabetes dataset
diabetes_data = pd.read_csv("/content/drive/MyDrive/diabetes.csv")

# Define the independent and dependent variables
X = diabetes_data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = diabetes_data['Outcome']

# Fit a multiple regression model
multiple_regression_model = sm.OLS(y, sm.add_constant(X)).fit()

# Print the summary of the multiple regression model
print(multiple_regression_model.summary())
```

```python
# prompt: Write a Pandas program to group by the first column and get second column as lists in rows Sample data: Original DataFrame col1 col2 0 C1 1 1 C1 2 2 C2 3 3 C2 3 4 C2 4 5 C3 6 6 C2 5 Group on the col1: col1 C1 [1, 2] C2 [3, 3, 4, 5] C3 [6] Name: col2, dtype: object

import pandas as pd

df = pd.DataFrame({'col1': [1, 1, 2, 3, 3, 4, 5, 6],
                   'col2': [4, 5, 6, 7, 8, 12, 1, 2]})

print("Original DataFrame:")
print(df)

grouped_df = df.groupby('col1')['col2'].apply(list).reset_index(name='list_col2')

print("\nGroup on the col1:")
print(grouped_df)
```
